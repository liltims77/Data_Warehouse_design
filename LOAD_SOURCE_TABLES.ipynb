```python
# Clean data: Normalize float_df column names 
float_df.columns = [col.strip().replace(" ", "_").lower() for col in float_df.columns]
```

```python
# Clean data: Normalize clickup_df column names 
clickup_df.columns = [col.strip().replace(" ", "_").lower() for col in clickup_df.columns]
```

```python
clickup_df
```

```python
# Create Dim_Team_Members

dim_team_members = float_df[['name', 'role']].drop_duplicates().reset_index(drop=True)
dim_team_members['Team_Member_ID'] = dim_team_members.index + 1
dim_team_members = dim_team_members[['Team_Member_ID', 'name', 'role']]
```

```python
dim_team_members
```

```python
# Create Dim_Projects
dim_projects = pd.concat([clickup_df[['project']], float_df[['project']]])
dim_projects = dim_projects.drop_duplicates().reset_index(drop=True)
dim_projects['Project_ID'] = dim_projects.index + 1
dim_projects = dim_projects[['Project_ID', 'project']]
```

```python
dim_projects
```

```python
# Create Dim_Date (Date Dimension)
date_range = pd.date_range(start='2022-01-01', end='2023-12-31')
dim_date = pd.DataFrame({
    'Date_ID': range(1, len(date_range) + 1),
    'Date': date_range,
    'Month': date_range.month,
    'Year': date_range.year
})
```

```python
dim_date
```

```python

# Merge datasets to create Fact_Allocations Join ClickUp and Float on Name and Project to combine tracked(hours) and allocated hours(estimated_hours)
merged_data = pd.merge(clickup_df, float_df, on=['name', 'project'], how='inner')
fact_allocations = merged_data.groupby(['name', 'project', 'date'], as_index=False).agg({
    'hours': 'sum',  # this is from ClickUp (Tracked Hours)
    'estimated_hours': 'sum'  # this is rom Float (Allocated Hours)
}).rename(columns={'hours': 'Tracked_Hours', 'estimated_hours': 'Allocated_Hours'})
```

```python
merged_data
```

```python
fact_allocations
```

```python
# Map IDs to the fact table
fact_allocations['Fact_ID'] = range(1, len(fact_allocations) + 1)
fact_allocations = pd.merge(fact_allocations, dim_team_members, on='name', how='left')
fact_allocations = pd.merge(fact_allocations, dim_projects, on='project', how='left')
fact_allocations = pd.merge(fact_allocations, dim_date, left_on='date', right_on='Date', how='left')
```

```python
fact_allocations = fact_allocations[['Fact_ID','Team_Member_ID', 'Project_ID', 'Date_ID', 'Tracked_Hours', 'Allocated_Hours']]
```
